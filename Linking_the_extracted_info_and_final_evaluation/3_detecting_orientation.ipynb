```python
import pickle
import os
import pandas as pd
```

```python
## Choose pred_materials.pkl for producing composition_orientation_pred.pkl, and gold_materials.pkl for composition_orientation_gold.pkl
res_materials = pickle.load(open(f'pred_materials.pkl', 'rb'))
# res_materials = pickle.load(open(f'gold_materials.pkl', 'rb'))
```

```python
flattened_list = [tup for sublist in res_materials for tup in sublist]
# flattened_list
```

```python
from collections import defaultdict

def detect_table_orientation(composition):
    orientation_results = {}

    # Group materials by PII and T_IDX
    grouped_materials = defaultdict(list)
    
#     for composition in materials:
    for entry in composition:
        PII_TID = entry[0].split('_')[:2]  # Get PII and T_IDX
        grouped_materials[tuple(PII_TID)].append(entry)
        
#     print(grouped_materials)

    for key, entries in grouped_materials.items():
        row_indices = [int(entry[0].split('_')[2]) for entry in entries]
        col_indices = [int(entry[0].split('_')[3]) for entry in entries]

        # Count occurrences of each row and column index
        row_count = defaultdict(int)
        col_count = defaultdict(int)
        

        for row in row_indices:
            row_count[row] += 1
        for col in col_indices:
            col_count[col] += 1
            
#         print(row_count, col_count)
            

        # Determine dominant row and column
        dominant_row = max(row_count, key=row_count.get)
        dominant_col = max(col_count, key=col_count.get)
        

        # Compare counts to determine orientation
        if row_count[dominant_row] > col_count[dominant_col]:
            orientation_results[key] = "Row oriented"
        else:
            orientation_results[key] = "Column oriented"

    return orientation_results
```

```python
# Detect orientation
orientation_discomat = detect_table_orientation(flattened_list)
# print(orientation_discomat)
```

```python
## Comment the appropirate one - very important - else you will get file missing later
pickle.dump(orientation_discomat, open(f'composition_orientation_pred.pkl', 'wb'))
# pickle.dump(orientation_discomat, open(f'composition_orientation_gold.pkl', 'wb'))
```
