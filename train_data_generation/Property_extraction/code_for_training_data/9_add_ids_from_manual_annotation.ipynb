```python
from collections import defaultdict
import os
import pickle
import re
import sys

import numpy as np
import pandas as pd
```

```python
data_dir = '../data/'

# all_train_data_pre_tuple_new - remember this new
comp_data = pickle.load(open(os.path.join(data_dir, 'all_train_data_pre_tuple_new.pkl'), 'rb')) #+ \
            #pickle.load(open(os.path.join(val_test_table_dir, 'all_val_test_data.pkl'), 'rb'))

##Loading the expert-annotated material id labelled tables
full_comp_data = pickle.load(open(os.path.join(data_dir, 'train_data_with_only_id_anno.pkl'), 'rb')) #+ \
            #pickle.load(open(os.path.join(val_test_table_dir, 'val_test_data.pkl'), 'rb')) 

# comp_data = pickle.load(open(os.path.join(table_dir, 'train_data.pkl'), 'rb')) + \
#             pickle.load(open(os.path.join(table_dir, 'val_test_data.pkl'), 'rb'))
# comp_train_data = pickle.load(open(os.path.join(table_dir, 'val_test_data.pkl'), 'rb'))
comp_data_dict = {(c['pii'], c['t_idx']): c for c in comp_data}
full_comp_data_dict = {(c['pii'], c['t_idx']): c for c in full_comp_data}
train_val_test_split = pickle.load(open(os.path.join(data_dir, 'train_val_test_split.pkl'), 'rb'))
splits = ['train', 'val', 'test']
```

```python
# train_data_before_tuple.pkl - manually annotated pkl

#copying ids from discomat
couu_col, couu_row, couu_null = 0, 0, 0
for pii_tid in comp_data_dict:
    if pii_tid in full_comp_data_dict:
        full_row_label = full_comp_data_dict[pii_tid]['row_label']
        full_col_label = full_comp_data_dict[pii_tid]['col_label']
        assert not(3 in full_col_label and 3 in full_row_label)
        if 3 in full_col_label:
            couu_col+=1
            #display(pd.DataFrame(full_comp_data_dict[pii_tid]['act_table']))
            ind = full_col_label.index(3)
            comp_data_dict[pii_tid]['col_label'][ind] = 3
        elif 3 in full_row_label:
            couu_row+=1
            ind = full_row_label.index(3)
            comp_data_dict[pii_tid]['row_label'][ind] = 3
    else:
        couu_null+=1
#         print(pii_tid)
#         print(comp_data_dict[pii_tid])
#         display(pd.DataFrame(comp_data_dict[pii_tid]['act_table']))
```

```python
all_train_data = []
all_val_test_data = []
for pii_tid, table in comp_data_dict.items():
    if (table['pii'], table['t_idx']) in train_val_test_split['train']:
        all_train_data.append(table)
    elif (table['pii'], table['t_idx']) in train_val_test_split['val']+train_val_test_split['test']:
        all_val_test_data.append(table)
    else:
        assert False
print(len(all_train_data))
print(len(all_val_test_data))
```

```python
pickle.dump(all_train_data, open(os.path.join(data_dir, 'all_train_data_pre_tuple_with_ids.pkl'), 'wb'))
```
