```python
import ast
import os
import sys
import numpy as np
import pandas as pd
from tqdm import tqdm
import pickle
```

```python
def changer(x):
    new = []
    for idx, num in enumerate(x):
        if num==21:
            new.append(17)
        elif num>21:
            new.append(num-1)
        else:
            new.append(num)
    return new
```

```python
table_dir = '../data/'
# assert os.path.exists(table_dir)
### update glass ids in all (glass id, prop name, value, unit) tuples under the key table["prop_tuples"]
all_train_data = pickle.load(open(os.path.join(table_dir, 'final_train_data_with_id_with_tuple.pkl'), 'rb'))
# all_val_data = pickle.load(open(os.path.join(table_dir, 'final_val_test_data_with_id_with_tuple.pkl'), 'rb'))
```

```python
prop_ids = [510, 1140, 2010, 2051, 540, 50, 180, 60, 160, 1014, 1015, 3012, 3174, 1116, 1119, 1020, 1011, 1118, 70, 1306]
prop_names = ['Density', 'GlassTransitionTg', 'RefractiveIndex', 'AbbeValue', 'YoungsModulus', 'ShearModulus', 'VickersHardness', 'PoissonRatio', 'FractureToughness', 'CrystallizationTemp', 'MeltingTemp', 'ElectricConduct', 'DielectricConst', 'TSofP', 'TAnnealingP', 'ExpansionCoeff', 'LiquidusTemperature', 'SofP', 'BulkModulus', 'ActivationEnergy']
```

```python
# convert 21 to 17
newtraindata = []
for idx, table in enumerate(all_train_data):
    if table['prop_table']:
        rl = table['row_label']
        cl = table['col_label']
        if 21 in rl or 21 in cl:

            table['prop_names'].remove('SofP')
            table['prop_names'].add('TSofP')

            # print(table['prop_row_col_indexes']) #add the 21st element to 17th elemnt, delete the 21st element later for all
            table['prop_row_col_indexes'][13] += table['prop_row_col_indexes'][17]

            msg = str(table['prop_tuples'][13])
            msg = msg.replace('Softening Point (Viscosity)', 'Softening Point (Temperature)')
            msg = ast.literal_eval(msg)
            del table['prop_tuples'][17]
            table['prop_tuples'][13] = msg

        table['row_label'] = changer(table['row_label'])
        table['col_label'] = changer(table['col_label'])

        del table['prop_row_col_indexes'][17]
    newtraindata.append(table)
```

```python
with open('../data/new_train_prop_data.pkl', 'wb') as f:
    pickle.dump(newtraindata, f)
f.close()

# with open('../data/new_val_test_prop_data.pkl', 'wb') as f:
#     pickle.dump(newvaldata, f)
# f.close()
```
